### Data scrapping from Lib.ru wesite

- Website is taking to long to respond for some links
- Links are given manually to acces the pages (links) of website

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
driver.get('http://az.lib.ru/janr/index_janr_1-25.shtml')
```

```python

dd_tag = driver.find_elements(By.TAG_NAME, 'dd')
dl_tags = dd_tag[1].find_elements(By.TAG_NAME, 'dl')

href_list = []
for dl in dl_tags:
    
    dt_tag = dl.find_element(By.TAG_NAME, 'dt')
    li_tag = dt_tag.find_element(By.TAG_NAME, 'li')
    a_tags = li_tag.find_elements(By.TAG_NAME, 'a')
    if len(a_tags) > 1:
        href = a_tags[1].get_attribute('href')
        href_list.append(href)

print(href_list)
```

```python
len(href_list)
```

```python
import os
import time

driver = webdriver.Chrome()
output_folder = r"D:\IITGN\NLP\Lib.ru"
os.makedirs(output_folder, exist_ok=True)

for index, url in enumerate(href_list, start=4801):
    try:
      
        driver.get(url)
        time.sleep(3)

        # Find the <noindex> tag
        noindex_tag = driver.find_element(By.TAG_NAME, "noindex")

        # Find the <div> tag inside the <noindex> tag
        div_tag = noindex_tag.find_element(By.TAG_NAME, "div")

        # Find all <dd> tags inside the <div> tag
        dd_tags = div_tag.find_elements(By.TAG_NAME, "dd")

        # Extract text from all <dd> tags
        dd_texts = [dd.text.strip() for dd in dd_tags]
        full_text = "\n".join(dd_texts)
        file_name = f"{index}.txt"
        file_path = os.path.join(output_folder, file_name)

        with open(file_path, "w", encoding="utf-8") as file:
            file.write(full_text)
        print(f"Extracted text from URL {url} and saved it to {file_path}")

    except Exception as e:
        print(f"Failed to extract data from {url}. Error: {e}")

```
